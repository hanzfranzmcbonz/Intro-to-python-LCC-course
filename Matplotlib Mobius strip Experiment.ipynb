```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation

# Mobius Strip parameters
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(-1, 1, 100)
u, v = np.meshgrid(u, v)

# Mobius Strip equations
x = (1 + (v / 2) * np.cos(u / 2)) * np.cos(u)
y = (1 + (v / 2) * np.cos(u / 2)) * np.sin(u)
z = (v / 2) * np.sin(u / 2)

# Create 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
mobius_strip = ax.plot_surface(x, y, z, cmap='viridis')

# Function to update the plot during animation
def update(frame):
    ax.view_init(elev=40, azim=frame)

# Create an animation
animation = FuncAnimation(fig, update, frames=np.arange(0, 360, 2), interval=1)

# Display the animation
plt.show()
```
